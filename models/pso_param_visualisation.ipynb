```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
df_cso_params = pd.read_csv('logs/params/pso/pso_20230429-104013.csv')
```

```python
columns = ["epoch", "pop_size", "c1", "c2", "w_min", "w_max", "score"]
data_raw = []
for item in df_cso_params.values:
    item_test = json.loads(item[2].replace("\'", "\""))
    tmp = []
    for k in columns[:-1]:
        tmp.append(item_test[k])
        
    tmp.append(item[6])
    data_raw.append(tmp)
    
data_raw.sort(key=lambda x: x[3], reverse=True)
pd.DataFrame(data_raw, columns=columns)
```

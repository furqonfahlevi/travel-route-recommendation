```python
def create_df(df_all, df_all_name):
    df_all_sub_name = [key_node, key_day, key_time, key_score] #key_fitness
    tmp_col = []
    tmp_sub_col = []
    for n in df_all_name:
        for sn in df_all_sub_name:
            tmp_col.append(n)
            tmp_sub_col.append(sn)

    col_df_all = [
        tmp_col,
        tmp_sub_col
    ]
    tmp_df_all = []
    for v in df_all[0]["max_tour"]:
        tmp_df_all.append([])

    for i, v in enumerate(df_all, start=1):
        for a, b in enumerate(df_all_sub_name):
            for t in range(len(tmp_df_all)):
                tmp_df_all[t].append(v[b][t])

    df_res = pd.DataFrame(tmp_df_all, columns=col_df_all, index=list(df_all[0]["max_tour"].values))
    df_res.index.name = "max_tour"
    display(df_res)
    display(df_res.describe())
    return df_res
```

```python
df_all = [df_bf_p1_c1_t1, df_pso_p1_c1_t1]
df_all_name = ["bf", "pso"]
df_res = create_df(df_all, df_all_name)
```

```python
def plot_result(data1, data2, key, ylabel, title):
    data_node = []
    node_columns = [data1[1], data2[1]]
    for i, d1, d2 in zip(
        data1[0]['max_tour'].values, 
        data1[0][key].values,
        data2[0][key].values,
    ):
        data_node.append([i,d1,d2])

    df_node = pd.DataFrame(data_node, columns=["max_tour"]+node_columns)
    df_node.plot(x='max_tour', y=node_columns, kind='line', figsize=(8 ,5))

    plt.title(title)
    plt.xlabel('Number of Destination (Input)')
    plt.ylabel(ylabel)
    plt.legend(node_columns)
    plt.show()
```

```python
plot_result([df_bf_p1_c1_t1, "Brute Force"], [df_pso_p1_c1_t1,"Particle Swarm Optimization"], key_node, "Number of Destination (Output)", "Number of Destination (Output) of PSO vs. BF with Criteria w1=1, w2=1, w3=1")
```

```python
plot_result([df_bf_p1_c1_t1, "Brute Force"], [df_pso_p1_c1_t1,"Particle Swarm Optimization"], key_time, "Time (s)", "Running Time of PSO vs. BF with Criteria w1=1, w2=1, w3=1")
```

```python
plot_result([df_bf_p1_c1_t1, "Brute Force"], [df_pso_p1_c1_t1,"Particle Swarm Optimization"], key_score, "Score", "Score of PSO vs. BF with Criteria w1=1, w2=1, w3=1")
```

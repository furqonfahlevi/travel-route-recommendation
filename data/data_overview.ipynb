# Distance Matrix

note: urutan wisata pada <b>data_distance_matrix</b> itu sama seperti urutan yang ada di <b>data_tourist_attractions</b>

````python
# Distance Matrix
with open("distance_matrix_tour.json", 'r') as f:
    data_distance_matrix = json.load(f)
    
print(type(data_distance_matrix))
print(len(data_distance_matrix))
````

````python
data_distance_matrix[0][1]
````

````python
pd.DataFrame(data_distance_matrix[0]).head()
````

````python
pd.DataFrame(data_distance_matrix[0])['distance_matrix'][1]
````

````python
temp_results = list()
for i in data_distance_matrix:
    temp_results2 = list()
    for x in i:
        if x['distance_matrix'] != None:
            x = x['distance_matrix']
            temp_results2.append(f"{x['rows'][0]['elements'][0]['distance']['text']} / {x['rows'][0]['elements'][0]['duration']['text']}")
        else:
            temp_results2.append(None)
    temp_results.append(temp_results2)
    
pd.DataFrame(temp_results, columns=places_name, index=places_name).head()
````
